```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.rc('axes', unicode_minus = False)

import pandas as pd
df = pd.read_csv('data/score.csv')
df
```

```python
#3)SW특기별 학생수 원그래프
group = df.groupby('SW특기').size()
plt.figure(figsize=(10,5))
plt.pie(group.values,labels=group.index, 
        autopct='%.1f%%', pctdistance=0.8,
        textprops={'size':10},
        wedgeprops={'width':0.4,'edgecolor':'w', 'linewidth':8})
plt.show()
```

```python
enroll = pd.read_csv('data/수강.csv')
enroll.head()
```

```python
group=enroll.groupby('강좌번호').size().to_frame('인원수')
cou = pd.read_csv('data/강좌.csv')
merge = cou.merge(group, on='강좌번호')
merge
```

```python
labels = merge['강좌이름']
values = merge['인원수']
```

```python
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values, labels=labels, wedgeprops={'width':0.55, 'edgecolor':'w', 'linewidth':6},
        colors=colors, pctdistance=0.7,
         autopct='%.0f%%')
plt.show()
```

```python
def grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'
```

```python
enroll = pd.read_csv('data/수강.csv')
enroll['학점'] = enroll['점수'].apply(grade)
enroll
```

```python
group = enroll.groupby('학점').size()
group
```

```python
plt.pie(group.values, labels=group.index + '학점', wedgeprops={'width':0.55, 'edgecolor':'w', 'linewidth':6},
        textprops={'color':'white', 'size':8},
        pctdistance=0.7, autopct='%.1f%%')
plt.show()
```
