```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt

#한글설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
df
```

```python
import numpy as np
def graph(name):
    filt = df['이름'] == name
    df2 = df[filt]
    if len(df2) == 0:
        return np.nan
    else:
        df3 = df2.loc[:, ['국어','과학','사회']]
        df4 = df3.iloc[0]
        plt.bar(df4.index, df4.values)
        return plt
image=graph('변덕규')
if pd.isnull(image):
    print('없음')
else:
    plt.show()

```

```python
name = list(df['이름'])
height = df['키']
kor = df['국어']
eng = df['영어']
mat = df['수학']
sci = df['과학']
soc = df['사회']
avg = list((kor+eng+mat+sci+soc)/5)
avg, name
```

```python
#x축 데이터가 많을 경우(그래프가 가로로 너무 길어지는 것을 방지)
plt.barh(name, height)
plt.xlabel('키')
plt.ylabel('학생이름')
plt.xlim(160,210)
plt.show()
```

```python
plt.figure(figsize=(10,5))
bar=plt.bar(name, height, width=0.7, color='none', ec='green')
plt.ylim(160,210)
bar[0].set_hatch('/')
for idx, h in enumerate(height):
    plt.text(idx, h+1, h, ha='center')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(name, avg)
plt.title('학생별 평균')
plt.ylim(0,100)
plt.yticks([y for y in range(0, 101, 5)])
plt.xlabel('학생이름')
plt.ylabel('평균점수')
for idx, a in enumerate(avg):
    plt.text(name[idx],avg[idx]+1, a, ha='center')
plt.show()
```
