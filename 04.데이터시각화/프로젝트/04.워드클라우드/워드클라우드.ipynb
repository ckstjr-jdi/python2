```python
with open('영어가사.txt', 'r') as file:
    text = file.read()
print(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud().generate(text)
```

```python
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.savefig('yellow가사.png')
plt.show()
```

```python
import numpy as np
from PIL import Image
```

```python
heart_mask = np.array(Image.open('heart_mask.png'))
```

```python
wc = WordCloud(width=800, 
               height=800, 
               background_color="#fdfeeb22",
               min_font_size=20,
               mask=heart_mask).generate(text)
```

```python
plt.Figure(figsize=(5,5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('yellow하트.png')
plt.show()
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
okt = Okt()
```

```python
with open('애국가.txt', 'r', encoding='utf-8') as file:
    text = file.read()
text
```

```python
okt = Okt()
nouns = okt.nouns(text)
nouns = [noun for noun in nouns if len(noun) > 1]
nouns
```

```python
counter = Counter(nouns)
counter
```

```python
wc = WordCloud(width=800, 
               height=800, 
               background_color="#eef3ffff",
               min_font_size=20,
               font_path='BATANG.TTC',
               max_font_size=250,
               mask=heart_mask).generate_from_frequencies(counter)
```

```python
plt.Figure(figsize=(5,5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('애국가.png')
plt.show()
```
