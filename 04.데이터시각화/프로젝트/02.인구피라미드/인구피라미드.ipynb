```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.rc('axes', unicode_minus = False)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
df_m15 = pd.read_excel('201508.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y')   
df_m15 = df_m15.head(1)
df_m15.iloc[0] = df_m15.iloc[0].str.replace(',','').astype(int).astype(int)
df_m15
```

```python
df_w15 = pd.read_excel('201508.xlsx', skiprows=3, index_col='행정기관', usecols='B, AB:AV')   
df_w15 = df_w15.head(1)
df_w15.iloc[0] = df_w15.iloc[0].str.replace(',','').astype(int).astype(int)
df_w15.columns=df_m15.columns
df_w15
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m15.columns, -df_m15.iloc[0]//1000, label='M')
plt.barh(df_w15.columns,df_w15.iloc[0]//1000, label='W')
plt.legend()
plt.xlabel('인구수(천단위)',loc='right')
plt.title('2015 대한민국 성별 인구 피라미드')
plt.savefig('2015년.png')
plt.show()
```
