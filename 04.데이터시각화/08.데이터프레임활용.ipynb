```python
import pandas as pd
df = pd.read_csv('c:/python/04.데이터시각화/data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt

#한글설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
df
```

```python
#1 학생별 키
name = df['이름']
height = df['키']
plt.figure(figsize=(10,3))
plt.ylim(150, 210)
plt.bar(name, height, color='indigo', hatch='.', ec='white')
plt.xticks(name, rotation=45, size=8)
for idx, h in enumerate(height):
    plt.text(idx, h+1 ,h, ha='center', size=8)
```

```python
#2 학생별 평균
df['평균'] = df.apply(lambda row:row['국어':'사회'].mean(), axis=1)
name = df['이름']
avg = df['평균']
plt.figure(figsize=(10,3))
plt.ylim(0, 100)
plt.bar(name, avg, color='darkgreen', hatch='.', ec='white')
plt.xticks(name, rotation=45, size=8)
for idx, h in enumerate(avg):
    plt.text(idx, h+1 ,f'{h:.2f}', ha='center', size=8)
```

```python
#3 학교별 평균키
group = df.groupby('학교')['키'].mean()
labels = group.index
values = group.values
plt.figure(figsize=(10,5))
plt.ylim(160, 200)
plt.bar(labels, values, color='olive', hatch='.', ec='white', width=0.5)
plt.xticks(labels, rotation=45, size=12)
for idx, value in enumerate(values):
    plt.text(idx, value+1 ,f'{value:.2f}cm', ha='center', size=8)
```

```python
#4학교별 평균점수
df['평균'] = df.apply(lambda row:row['국어':'사회'].mean(), axis=1)
group = df.groupby('학교')['평균'].mean()
labels = group.index
values = group.values
plt.figure(figsize=(10,5))
plt.ylim(0, 100)
plt.bar(labels, values, color='darkorange', hatch='.', ec='white', width=0.5)
plt.xticks(labels, rotation=45, size=12)
for idx, value in enumerate(values):
    plt.text(idx, value+1 ,f'{value:.2f}점', ha='center', size=8)
```

```python
#5.SW특기별인원수
df['SW특기'] = df['SW특기'].str.capitalize()
group = df.groupby('SW특기').size()
labels = group.index
values = group.values
plt.figure(figsize=(10,5))
plt.ylim(0, 6)
plt.bar(labels, values, color='orangered', hatch='.', ec='white', width=0.5)
plt.xticks(labels, rotation=45, size=12)
for idx, value in enumerate(values):
    plt.text(idx, value+0.02 ,f'{value}명', ha='center', size=10)
```

```python
#학생별(이름) 평균점수
enroll = pd.read_csv('data/수강.csv')
group=enroll.groupby('학생번호')['점수'].mean().to_frame('평균점수')
stu = pd.read_csv('data/학생.csv')
merge = stu.merge(group, how='left', on='학생번호')
merge.fillna({'평균점수':0}, inplace=True)
```

```python
#학생별 평균점수 막대그래프
nos = merge['학생번호']
name = merge['학생이름']
grades = merge['평균점수']
xticks = [f'{nos[idx]}\n{name[idx]}' for idx in range(len(nos))]
plt.bar(name, grades)
plt.ylim(0,100)
plt.xticks(name, xticks, rotation=45)
for idx, grade in enumerate(grades):
    plt.text(idx, grade+1, f'{grade:.2f}', ha='center')
```

```python
#강좌별(이름) 평균
enroll = pd.read_csv('data/수강.csv')
group=enroll.groupby('강좌번호')['점수'].mean().to_frame('평균점수')
cou = pd.read_csv('data/강좌.csv')
merge = cou.merge(group, how='left', on='강좌번호')
merge.fillna({'평균점수':0}, inplace=True)
merge
```

```python
#강좌별(이름) 평균 막대그래프
nos = merge['강좌번호']
names = merge['강좌이름']
grades = merge['평균점수']
xticks = [f'{name[idx]}({nos[idx]})' for idx in range(len(nos))]
plt.bar(names,grades)
plt.ylim(0,100)
plt.xticks(names, xticks, rotation=45)

yticks = [y for y in range(0, 101, 20)]
plt.yticks(yticks, [f'{y}점' for y in yticks])
for idx, grade in enumerate(grades):
    plt.text(idx, grade+1, f'{grade:.2f}', ha='center')
```
